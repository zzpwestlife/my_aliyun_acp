### 1.4 Basic Conversation Call
Let's start with a simple conversation. The following code creates an assistant named "Company Assistant," that answers questions about company operations. Try it with the example question: "Choosing a project management tool"



````python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url=os.getenv("DASHSCOPE_API_BASE"),
)

def get_qwen_response(prompt):
    response = client.chat.completions.create(
        model="qwen-max",
        messages=[
            # The system message is used to set the role and task of the LLM
            {"role": "system", "content": "You are an assistant responsible for answering questions in a global educational content development company. Your name is Alex, and you're here to help your colleagues with their inquiries."},
            # The user message is used to input the user's question
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content

response = get_qwen_response("What tools should our company use for project management?")
print(response)
````

### 1.5 Multi-Round Conversations

In real-world applications, users rarely present complete questions all at once. Consider this scenario: A new employee first asks, "What project management tools does our company use?" After receiving the answer, they follow up with, "Then how do I apply for an account?" If the LLM cannot "remember" the previous conversation, it won't know which tool the user is asking about when it comes to the account application process.

This is where **multi-round conversations** come in: they enable the LLM to reference historical conversation information, understand contextual connections, and provide coherent and accurate responses.

#### How Multi-Round Conversations Work

In the OpenAI SDK, the key to implementing multi-round conversations lies in the **messages** parameter. This is a list that contains the complete history of the conversation. Each message has two essential fields:

- **role**: The role of the message, which can be `system` (system instructions), `user` (user input), or `assistant` (model responses)
- **content**: The specific content of the message

The LLM generates responses based on all messages in the messages list. Therefore, you need to preserve the entire conversation history‚Äîincluding both user questions and model responses‚Äîin this list.


````python
def multi_turn_chat():
    # Initialize the conversation history, starting with the system message
    conversation_history = [
        {"role": "system", "content": "Your role is to answer questions about the company. Your name is XiaoMi, and you assist colleagues with their inquiries."}
    ]
    
    # Multi-turn chat simulation
    user_questions = [
        "What tools should we use for project management at our company?",
        "How do I apply for an account for this tool?",
        "How long does it take for an application to be approved?"
    ]
    
    for question in user_questions:
        print(f"üë§ UserÔºö{question}")
        
        # Add the user question to the conversation history
        conversation_history.append({"role": "user", "content": question})
        
        # Call the large model, passing in the complete conversation history
        response = client.chat.completions.create(
            model="qwen-max",
            messages=conversation_history  # includes all the messages 
        )
        
        # Acquire the model reply
        assistant_message = response.choices[0].message.content
        print(f"ü§ñ Assistant: {assistant_message}\n")
        
        # Add the model reply to the conversation history, so it can be used for the next round of conversation.
        conversation_history.append({"role": "assistant", "content": assistant_message})

multi_turn_chat()
````

Running the code above, you'll find that the LLM can understand that "this tool" refers to the project management tool mentioned earlier, and answer questions about the account application and approval process in a targeted manner. This is the role of multi-round conversations.

> üí° **Tips**:
> - Conversation history occupies space in the **context window**. If there are too many conversation rounds, you may need to truncate or summarize the historical messages.
> - In production environments, conversation history usually needs to be persistently stored (e.g., in a database) to support continuous conversations across sessions for users.
> - To learn more about best practices for multi-round conversations, you can refer to the [Alibaba Cloud Bailian Multi-Round Conversation Documentation](https://www.alibabacloud.com/help/zh/model-studio/multi-round-conversation).

### 1.6 Streaming Output

After running the code above, you'll notice it takes about 20 seconds to receive the complete response. By default, the API waits until the model finishes generating before returning the full response at once. In real-world applications, this delay can negatively impact user experience‚Äîimagine users staring at a blank screen for 20 seconds!

Fortunately, you can use "streaming output" to address this issue. With streaming output, the model sends responses incrementally as it generates them, much like how humans type, so users can see partial results immediately. This significantly improves user experience. Next, let‚Äôs see how to implement streaming output.
> üí° Tip: Streaming output only changes the way content is displayed; the model's reasoning process and quality of the final answer remain unchanged. You can confidently use this feature to improve your application's user experience.

To implement streaming output, simply add the `stream=True` parameter to the previous code and adjust the output method:

````python
def get_qwen_stream_response(user_prompt, system_prompt):
    response = client.chat.completions.create(
        model="qwen-max",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        stream=True
    )
    for chunk in response:
        yield chunk.choices[0].delta.content

response = get_qwen_stream_response(user_prompt="What tools should our company use for project management?", system_prompt="You are an assistant responsible for answering questions in a global educational content development company. Your name is Alex, and you're here to help your colleagues with their inquiries.")
for chunk in response:
    print(chunk, end="")
````

### 2.1. Text Generation Workflow

Let‚Äôs look at how a large language model turns the partial input **"ACP is a very"** into a full sentence, step by step.

The text generation process of a LLM is divided into five stages: **Tokenization**, **Token Embedding**, **Model Inference**, **Decoding & Autoregression**, and **Output Text**.

**Stage 1: Tokenization**

Computers can‚Äôt work with raw text, so the first step is to convert our input text "ACP is a very" into a numerical format that the computer can process. This process is called **tokenization**.

A **token** is the basic unit obtained after the tokenizer encodes the text. Each token corresponds to a unique integer ID in the vocabulary. Note that **tokens are usually sub-word fragments or character fragments**; they are not necessarily complete "words," nor do they always carry standalone meaning. For example, an English word might be split into a root and a suffix, Chinese text might be split by character or common phrases, and spaces or punctuation may also be encoded into tokens.

With "ACP is a very" as our example, the tokenizer encodes it into several tokens and assigns an ID to each. To dive deeper into the tokenization details of Qwen, you can refer to its official documentation: [Tokenization Note](https://github.com/QwenLM/Qwen/blob/main/tokenization_note.md).

**Stage 2: Token Embeddings**

Now that we have a sequence of numeric IDs, the values themselves are not inherently meaningful (for example, ID 500 is not necessarily more "important" or "relevant" than ID 50). For the model to grasp the semantic representations of tokens, we need to convert these discrete IDs into mathematical representations containing rich semantic information‚Äî**Vectors**.

This conversion is done via a massive table known as an **Embedding Matrix**. Essentially, the system uses the token ID to "look up" the corresponding row in this table. Each token's vector contains its coordinates in a multi-dimensional semantic space.

However: word order is crucial ("dog chases cat" is very different from "cat chases dog"). However, the standard transformer model has no built-in sense of word order. Therefore, **Positional Encoding** must be added to each token vector, allowing the model to distinguish that "dog" is the first word, "cat" is the third, and so on.

**Stage 3: Model Inference**

At this point, the vector sequence carrying both semantic and positional information is fed into the **Decoder Blocks** of the Transformer model for calculation. This process is known as the **Forward Pass**.

The vector sequence passes layer by layer through dozens of structurally similar decoder blocks. Using **Causal Self-Attention** and **Feed-Forward Networks**, the model repeatedly compresses and refines the information. This produces a **Hidden State** vector corresponding to the last token (in out example, "very"). Think of this vector as the model's **compressed prediction**‚Äîa distilled representation of what should follow "ACP is a very". This vector is then projected onto the full vocabulary through a linear layer, producing a **Score Vector (Logits)** with one entry for every token.

**Stage 4: Decoding & Autoregression**

Once we have the logits, the LLM uses the softmax function to convert these scores into a probability distribution $P(\text{next\_token} \mid \text{context})$, representing how likely each token is to come next. Finally, the model picks which token to output based on a **Decoding Strategy**. Decoding strategies fall into two main categories:

- **Approximately Deterministic Decoding Methods:** These include Greedy Decoding (selecting the token with the highest probability each time) and Beam Search (tracking multiple candidate paths).
- **Stochastic Sampling Decoding Methods:** These include Top-p (Nucleus Sampling) and Top-k Sampling, which randomly select from a high-probability candidate set.

Many online services use stochastic sampling by default, which explains why you might get "slightly different answers" even if the input is identical.

By adjusting the **temperature**, you can tweak the "sharpness" of the softmax output probability distribution. A lower temperature increases the probability differences between tokens, concentrating the distribution on a few high-probability tokens, making the output more deterministic. Meanwhile, a higher temperature flattens the distribution, increasing the diversity of the model's output. Combined with parameters like **top_p** (which controls the range of the candidate set for sampling), you can fine-tune the trade-off between creativity and consistency of the model's output. More on these parameters in the next section.

For example, the table below shows some candidate tokens and their probabilities (the actual vocabulary is much larger):

| Candidate Token | Probability |
|-----------------|-------------|
| informative     | 50%         |
| fun             | 25%         |
| enlightening    | 20%         |
| boring          | 5%          |

Once the next token is selected (for example, the model chooses "informative"), this newly generated token is **appended** to the end of the original input sequence to form a new input: "ACP is a very informative". The model then repeats Stage 3 and Stage 4 based on this new sequence to predict the subsequent token. This word chain process‚Äîwhere each new word links to the previous ones‚Äîis called **Autoregressive Generation**.

The model's autoregressive loop continues until it hits a stopping condition:
- A special **End-of-Sequence (EOS) token** is generated. 
- The pre-set maximum generation length is reached.
- A user-specified sequence of stop words is generated.

### 2.2 Parameters affecting the randomness of content generation in LLMs

Consider a question-and-answer scenario where the user asks: "What can you learn in the ACP course on LLMs?" To simulate the content generation process of an LLM, we have pre-defined a candidate token set consisting of the following tokens: "**RAG**", "**prompt**", "**model**", "**writing**", and "**drawing**". The LLM will select one of these five candidate tokens as the output (next-token), as shown below.
> User question: What can I learn in the ACP course?<br><br>
> LLM response: RAG <br>

In this process, two important parameters influence the LLM's output: temperature and top_p. These parameters control the randomness and diversity of the content generated by the LLM. Below, we introduce how these two parameters work and how to use them.

#### 2.2.1 Temperature

Before generating the next token, the LLM first calculates an initial probability distribution for the candidate tokens. This distribution represents the probability of each candidate token being selected as the next-token. Temperature acts as a regulator that alters the probability distribution of the candidate tokens, influencing the LLM's content generation. By adjusting this parameter, you can flexibly control the diversity and creativity of the text generated.

To better understand, the figure below illustrates the impact of different temperature values on the probability distribution of the candidate tokens. The plotting code is located in the /resources/2_1 directory.

<a href="https://img.alicdn.com/imgextra/i3/O1CN01lvmaaI1nrY4zNhZIp_!!6000000005143-2-tps-3402-1228.png" target="_blank">
<img src="https://img.alicdn.com/imgextra/i3/O1CN01lvmaaI1nrY4zNhZIp_!!6000000005143-2-tps-3402-1228.png" width="1000">
</a>

The low, medium, and high temperatures settings in the figure are based on the range [0, 2) of the Qwen-Max model.

As shown in the figure, as the temperature increases from low to high (0.1 -> 0.7 -> 1.2), the probability distribution becomes flatter. The probability of the candidate token "RAG" decreases from 0.8 -> 0.6 -> 0.3. Although it remains the most probable token, its probability is now closer to those of other candidate tokens. Consequently, the final output transitions from relatively fixed to increasingly diverse.

Depending on the use case, consider the following recommendations for setting the temperature parameter:
- Clear answers (such as, generating code): Lower the temperature.
- Creative variety (such as, ad copy): Increase the temperature.
- No special requirements: Use the default temperature (usually within the medium range).

Note that when temperature = 0, although randomness is minimized, it does not guarantee identical outputs every time. For a deeper understanding, refer to [the underlying algorithm implementation of temperature](https://github.com/huggingface/transformers/blob/v4.49.0/src/transformers/generation/logits_process.py#L226).

Next, let‚Äôs experience the effect of temperature. By adjusting the temperature value, ask the same question 10 times and observe the fluctuations in the responses.
> The example code for temperature is similar to the one used in the explanation of top_p below, so it has been encapsulated for subsequent use.

````python
import time

def get_qwen_stream_response(user_prompt, system_prompt, temperature, top_p):
    response = client.chat.completions.create(
        model="qwen-max",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=temperature,
        top_p=top_p,
        stream=True
    )

    for chunk in response:
        yield chunk.choices[0].delta.content


# Prints streaming output from Qwen-Max with optional parameters
def print_qwen_stream_response(
    user_prompt,
    system_prompt,
    temperature=0.7,
    top_p=0.8,
    iterations=10
):
    for i in range(iterations):
        print(f"Output {i + 1}: ", end="")
        # Delay to avoid hitting API rate limits
        time.sleep(0.5)
        response = get_qwen_stream_response(user_prompt, system_prompt, temperature, top_p)
        output_content = ''
        for chunk in response:
            if chunk:  # Avoid NoneType errors
                output_content += chunk
        print(output_content)


# Qwen-Max Model Info:
# - Temperature range: [0, 2), default: 0.7
# - Set temperature=0 for deterministic output
print_qwen_stream_response(
    user_prompt="A horse can also be referred to as",
    system_prompt="Continue this sentence with a phrase of five words or fewer.",
    temperature=0
)
````

````python
# Set temperature=1.9
print_qwen_stream_response(user_prompt="A horse can also be referred to as", system_prompt="Continue this sentence with a phrase of five words or fewer.", temperature=1.9)
````

The experiment clearly shows that as the temperature increases, the model generates increasingly varied and diverse content.

#### 2.2.2 top_p: Control the sampling range of the candidate token set

top_p (also known as nucleus sampling) is a filtering mechanism that selects a subset of candidate tokens whose cumulative probability meets certain threshold. Specifically, tokens are sorted in descending order of probability, and the smallest set of top-ranked tokens whose cumulative probability is **at least** the top-p threshold is selected as the new candidate set.

The figure below shows the sampling effect of different top_p values on the candidate token set.

<a href="https://img.alicdn.com/imgextra/i3/O1CN01UnTMMP20ZJJso60aG_!!6000000006863-2-tps-2568-1944.png" target="_blank">
<img src="https://img.alicdn.com/imgextra/i3/O1CN01UnTMMP20ZJJso60aG_!!6000000006863-2-tps-2568-1944.png" width="800">
</a>

In the illustration, the blue part represents tokens whose cumulative probability reaches the top_p threshold (such as, 0.5 or 0.8), forming a new candidate set; the gray part represents tokens that are not selected.

When top_p = 0.5, only the highest-probability token (‚ÄúRAG‚Äù) is included in the candidate set; when top_p = 0.8, the pool expands to include ‚ÄúRAG,‚Äù ‚Äúprompt,‚Äù and ‚Äúmodel,‚Äù from which the next token is sampled randomly.

In summary, the effect of the top_p parameter on LLM output can be described as follows:
- Higher top_p values: Broader candidate range, more diverse content, ideal for tasks like poetry, storytelling, or brainstorming.
- Lower top_p values: Narrower candidate range, more stable output, suitable for news drafts, code generation, and other scenarios requiring clear answers.
- Extremely low top_p values (such as, 0.0001): Theoretically, the model only selects the highest-probability token, resulting in very stable output. However, in practice, due to factors such as distributed systems and additional adjustments to model outputs, slight randomness may still be introduced, making it impossible to guarantee completely consistent output every time.


Next, explore the effect of top_p by running the same question 10 times with different top_p values and observing the variation in responses.


````python
# Qwen-Max model: The value range of top_p is (0,1], with a default value of 0.8.
# Set top_p=0.001
print_qwen_stream_response(
    user_prompt="Name an intelligent gaming smartphone, it could be",
    system_prompt="Suggest a short and catchy name for a smart gaming smartphone, using 4 words or fewer.",
    top_p=0.001
)
````

````python
# Set top_p=0.8
print_qwen_stream_response(
    user_prompt="Name an intelligent gaming smartphone, it could be", 
    system_prompt="Suggest a short and catchy name for a smart gaming smartphone, using 4 words or fewer.", 
    top_p=0.8
)
````

Based on the experimental results, it can be observed that the higher the top_p value, the greater the randomness in the output of LLMs.

#### 2.2.3 Summary

**Should temperature and top_p be adjusted simultaneously?**

While temperature and top_p influence randomness in different ways, they are commonly used together in practice. However, for beginners or during initial experimentation, it‚Äôs advisable to adjust one parameter at a time to better understand its individual effect. Once familiar with their behavior, combining both allows finer control over creativity and coherence.

><br>**Knowledge Extension: top_k**<br>In the Qwen series models, the parameter top_k also has capabilities similar to top_p. Refer to the [Qwen API Documentation](https://www.alibabacloud.com/help/en/model-studio/qwen-api-reference). Top_k is a sampling method that restricts the next token selection to the k most probable tokens, from which one is sampled according to the (optionally temperature-scaled) probabilities. Generally speaking, the larger the top_k, the more diverse the generated content; the smaller the top_k, the more fixed the content. When top_k is set to 1, the model only selects the Token with the highest probability, making the output more stable but also causing a lack of variation and creativity.<br><br>
>**Knowledge Extension: seed**<br>In the Qwen series models, the parameter seed also supports controlling the determinism of the generated content. Refer to the [Qwen API Documentation](https://www.alibabacloud.com/help/en/model-studio/qwen-api-reference). The seed parameter enables reproducible sampling. When the same seed, prompt, and generation parameters (e.g., temperature, top_p) are used, the model will typically produce identical outputs. However, full determinism is not always guaranteed due to factors such as hardware-level non-determinism (e.g., GPU floating-point operations) or backend implementation details.
<br><br>

**Why does randomness still exist when controlling the output of LLMs (LLM) by setting temperature, top_p, and seed?**

Even if temperature is set to 0, top_p is set very low(like 0.0001), and the same seed is used, the generated results for the same question may still be inconsistent. This is because some complex factors may introduce slight randomness, such as the LLM running in a distributed system, or optimization being applied to the LLM's output.

For example:
A distributed system is like slicing bread with different machines. Although each machine operates according to the same settings, subtle differences between devices may still result in slightly different slices of bread.  



## ‚öôÔ∏è 3. Enable LLMs to answer private knowledge questions
To enable LLMs to answer questions about internal or proprietary knowledge, you can choose one of the following approaches:
- **Context Engineering (No Model Changes)**<br>
    Provide the relevant internal knowledge directly within the prompt when making a request.
- **Model Customization (Fine-tuning or Training)**<br>
    Fine-tune or train a new model to better understand and respond to domain-specific queries.

Given the high cost and complexity of model training, for most internal knowledge Q&A use cases, it's often more efficient to include the necessary context in the prompt.

````python
# User's question
user_question = "I'm part of Software Team 1. What project management tool should we be using?"

# Internal knowledge about company tools
knowledge = """The company provides two main project management tools:
  
  1. **Jira**: Best suited for software development teams practicing agile methodologies like Scrum and Kanban. It offers robust features such as issue tracking, sprint planning, and time logging.
  
  2. **Microsoft Project**: Designed for enterprise-level or complex projects requiring detailed planning, resource allocation, and budget control. It's ideal for teams needing strict oversight over timelines and costs.
  
  Recommendation: While Microsoft Project is the company-standard solution and fully licensed, Software Teams 1, 3, and 4 are currently using Jira. There are plans to transition these teams to Microsoft Project by mid-2026.
"""

# Use a named assistant persona (for clarity and branding)
response = get_qwen_stream_response(
    user_prompt=user_question,
    system_prompt="You are Alex, an internal support assistant at an educational content development company. Use the provided context to answer the user‚Äôs question clearly and accurately." + knowledge,
    temperature=0.7,
    top_p=0.8
)

for chunk in response:
    print(chunk, end="")
````

When you provide reference information directly within a prompt, LLMs can answer questions involving private or internal knowledge. However, this approach has notable limitations. Prompts have maximum length constraints, and introducing large amounts of background data can lead to overly long inputs. Consequently, model efficiency may be reduced and token limits may be exceeded, preventing the model from generating a response.

To address these limitations, you can implement a system that automatically retrieves relevant internal knowledge when a user asks a question. The retrieved document snippets are then combined with the user‚Äôs query and fed into the LLM to generate a final answer. Such a system lets you avoid flooding the prompt with excessive background data. This approach is commonly referred to as **Retrieval-Augmented Generation (RAG)**.

A typical RAG pipeline consists of two main stages:

### 3.1 Indexing phase
<img src="https://img.alicdn.com/imgextra/i3/O1CN01h0y0Uy1WH30Q7FRDJ_!!6000000002762-2-tps-1592-503.png" width="1000">

During the indexing phase, internal knowledge documents or fragments are transformed into a format suitable for fast and accurate retrieval. This involves splitting the content into chunks, converting each chunk into a numerical representation (called an embedding) using specialized models, and storing these embeddings in a vector database.

This vectorization process preserves semantic meaning and enables efficient similarity searches across large knowledge bases. As a result, queries can quickly locate the most relevant documents, significantly boosting both the accuracy and generation speed of responses.

### 3.2 Retrieval and generation phase
<img src="https://img.alicdn.com/imgextra/i3/O1CN01JJuO2e1TvrTOk9Dw7_!!6000000002445-2-tps-2987-938.png" width="1000">

During the retrieval phase, the system searches the vector database for the most relevant document snippets based on the user‚Äôs query. These results are then combined with the original question and passed to the LLM, which generates a final, contextually informed answer. The LLM can then effectively respond to questions involving internal or proprietary knowledge without requiring full access to the underlying documents.

In summary, applications based on the RAG structure avoid various issues caused by inputting entire reference documents as background information while extracting the most relevant parts through retrieval, thus improving the accuracy and relevance of the LLM output.

The reasoning model [QwQ](https://www.alibabacloud.com/help/en/model-studio/models#fd7d313ca5khs) of Alibaba Cloud has strong reasoning capabilities. The model will first output the thought process and then provide the response content.  



````python
reasoning_content = ""  # Define the complete reasoning process
answer_content = ""     # Define the complete response
is_answering = False   # Determine whether the reasoning process has ended and the response has started

# Create a chat completion request
completion = client.chat.completions.create(
    model="qwq-plus",  # Here, qwq-plus is used as an example; you can replace it with another model name as needed
    messages=[
        {"role": "user", "content": "Which is larger, 9.9 or 9.11?"}
    ],
    stream=True,
    # Uncomment the following to return token usage in the last chunk
    # stream_options={
    #     "include_usage": True
    # }
)

print("\n" + "=" * 20 + " Reasoning Process " + "=" * 20 + "\n")

for chunk in completion:
    # If chunk.choices is empty, print usage
    if not chunk.choices:
        print("\nUsage:")
        print(chunk.usage)
    else:
        delta = chunk.choices[0].delta
        # Print the reasoning process
        if hasattr(delta, 'reasoning_content') and delta.reasoning_content != None:
            print(delta.reasoning_content, end='', flush=True)
            reasoning_content += delta.reasoning_content
        else:
            # Start responding
            if delta.content != "" and is_answering is False:
                print("\n" + "=" * 20 + " Complete Response " + "=" * 20 + "\n")
                is_answering = True
            # Print the response process
            print(delta.content, end='', flush=True)
            answer_content += delta.content
````
